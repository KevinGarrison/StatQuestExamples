```python
# Imports
import pandas as pd, numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the csv file into a pandas dataframe object
df = pd.read_csv('../Data/marathon_runners_pace.csv',usecols=['Gender', 'Pace (min/km)', 'Runner_ID'], sep=';', index_col='Runner_ID')
```

```python
# Look at the data
df
```

```python
# Plot the histograms for the data seperated by Gender including the density function
for bin_size in range(2, 20, 3):
    sns.histplot(data=df,x='Pace (min/km)', hue='Gender', bins=bin_size, kde=True)
    plt.xlabel('Pace')
    plt.ylabel('Pace Frequency')
    plt.title(f'Histogram of Observations Bin Size{bin_size}')
    plt.show()
```
